AR(1)
-----

```r
set.seed(1)
x<-w<-rnorm(100)
for (t in 2:100) x[t] <- 0.6*x[t-1] + w[t]
```

```r
layout(1:2)
plot(x,type='l')
acf(x)
```

Use the **ar** command to fit an autoregressive model to our simulated AR(1) proces. 
Use maximum likelihood estimation (MLE) as the fitting procedure

```r
x.ar <- ar(x, method="mle")
x.ar$order
```

The **ar** command has successfully determined that our underlying time series model is an AR(1) process. 

```r
x.ar$ar
```

The MLE procedure has produced an estimate **0.523**, which is slightly lower than the true value of **0.6**. 

Finally, we can use the standard error (with the asymptotic variance) to construct 95% confidence intervals around the underlying parameters. We simply create a vector c(-1.96,1.96) and then multiply it by the standard error.

```r
x.ar$ar + c(-1.96,1.96)*sqrt(x.ar$asy.var)
```

The true parameter does fall within the 95% CI. 

How about alpha1 = -0.6?

```r
set.seed(1)
x <- w <- rnorm(100)
for (t in 2:100) x[t] <- -0.6*x[t-1] + w[t]
layout(1:2)
plot(x, type="l")
acf(x)
```

```r
x.ar <- ar(x, method="mle")
x.ar$order
```

```r
x.ar$ar
```

```r
x.ar$ar + c(-1.96,1.96)*sqrt(x.ar$asy.var)
```

AR(2)
----

```r
set.seed(1)
x <- w <- rnorm(100)
for (t in 3:100) x[t] <- 0.666*x[t-1] -0.333*x[t-2] + w[t]
layout(1:2)
plot(x, type="l")
acf(x)
```

```r
x.ar <- ar(x, method="mle")
```

```r
x.ar$order
```

```r
x.ar$ar
```

Convergence warning message. Notice that R actually uses the **arima0** function to calculate the AR model. AR(p) models are actually ARIMA(p,0,0) models. 
Financial Data
-------------

Let's begin by obtaining the stock price for Amazon (AMZN) using **quantmod**: 

```r
library('TTR')
require('quantmod')
getSymbols('AMZN',src='yahoo')
AMZN[1:10]
```

```r
plot(Cl(AMZN))
```

Take the logarithmic returns of AMZN, then the first-order difference of the series. =>convert the original price series from a non-stationary series to a (potentially) stationary one. 

Ref: https://quantivity.wordpress.com/2011/02/21/why-log-returns/

```r
amznrt = diff(log(Cl(AMZN)))  #differenced log returns
plot(amznrt)
```

We want to know if the differenced series looks like white noise. If it does not then there is unexplained serial correlation, which might be "explained" by an autoregressive model. 

```r
acf(amznrt, na.action=na.omit)
```

We notice a statistically significant peak at k=2. Hence there is a reasonable possibility of unexplained serial correlation. Be aware though, that this may be due to sampling bias. As such, we can try fitting an AR(p) model to the series and produce confidence intervals for the parameters:

```r
amznrt.ar <- ar(amznrt, na.action=na.omit)
amznrt.ar$order
```

```r
amznrt.ar$ar
```

```r
amznrt.ar$asy.var
```

alpha1=-0.0213, alph2=-0.0618. We can find the 95% intervel:

```r
-0.0213 + c(-1.96,1.96)*sqrt(4.11e-4)
```

```r
-0.0618 + c(-1.96,1.96)*sqrt(4.11e-4)
```

We can see that for alpha1 zero is contained withing the CI, while for alpha2 zero is not contained in the CI. Hence we should be very careful in thinking that we really have an underlying generative AR(2) model for AMZN. 
S&P500 US Equity Index
---------------------

```r
getSymbols("^GSPC")
GSPC[1:10]
```

```r
plot(Cl(GSPC))
```

```r
gspcrt=diff(log(Cl(GSPC))) #the first order difference of the log closing prices
```

```r
plot(gspcrt)
```

It is clear from this chart that the volatility is not stationary in time. We can see the highly volatile period around 2008. This is also reflected in the plot of the correlogram.  

```r
acf(gspcrt, na.action=na.omit)
```

Ultimately we need a more sophisticated model than an autoregressive model of order p. However, we can still try fitting such a model.

```r
gspcrt.ar <- ar(gspcrt, na.action=na.omit)
gspcrt.ar$order
```

```r
gspcrt.ar$ar
```
